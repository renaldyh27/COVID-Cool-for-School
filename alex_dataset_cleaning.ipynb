```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
import os
```

```python
months = ["January", "February", "March", "April", "May", "June", "July"]
data = pd.read_csv("Datasets/border_crossings.csv")
data
```

```python
data = data.dropna(axis=0)
data
```

```python
for month in months:
    data[month] = data[month].str.replace(',', '').astype(int)
```

```python
data
```

```python
x = data.iloc[0]
x = list(x[months])
x
```

```python
plt.bar(months,x)
plt.title("Montly Arrival of Pedestrians at Andrade")
plt.xlabel("Month")
plt.ylabel("Number of People")
plt.show()
```
